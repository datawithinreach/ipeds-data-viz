# Urban Institute Education Data API — Endpoint Reference

The Urban Institute Education Data Portal is **free and requires no API key**.  
It aggregates data from IPEDS, College Scorecard, Federal Student Aid (FSA), CCD, CRDC, and other federal sources.

| Detail | Value |
|--------|-------|
| **Base URL** | `https://educationdata.urban.org/api/v1/` |
| **Docs** | https://educationdata.urban.org/documentation/ |
| **Sections** | `college-university`, `schools`, `school-districts` |
| **Institution ID** | `unitid` (IPEDS ID) for colleges; `ncessch` for schools; `leaid` for districts |

All endpoints follow the pattern: `{base}/{section}/{source}/{topic}/{year}/[sub-filters]`

---

https://educationdata.urban.org/api/v1/api-endpoints
Lists general information about each endpoint (e.g., their IDs, URLs, and years available).

https://educationdata.urban.org/api/v1/api-downloads
Lists information about downloadable data files and codebooks for each endpoint. All URLs to direct downloads follow this format: https://educationdata.urban.org/csv/{file_dir}/{file_name}.

https://educationdata.urban.org/api/v1/api-variables
Lists information about each variable in the portal (e.g., their data types and formats, labels, and special values). Each record is a unique variable.

https://educationdata.urban.org/api/v1/api-endpoint-varlist
Lists information about each variable in the portal, broken out by endpoint. Each record is a unique endpoint_id and variable combination.

```python
import requests
import pandas as pd

BASE = "https://educationdata.urban.org/api/v1"
COLLEGE = f"{BASE}/college-university"
SCHOOL = f"{BASE}/schools"
DISTRICT = f"{BASE}/school-districts"

SAMPLE_UNITID = 166027   # MIT
SAMPLE_YEAR = 2020
```

## Metadata Endpoints

These describe the portal itself and are useful for programmatic discovery.

| Endpoint | Description |
|----------|-------------|
| `GET /api/v1/api-endpoints` | Lists every endpoint (IDs, URLs, years available) |
| `GET /api/v1/api-downloads` | Downloadable CSV files and codebooks — download URLs follow `https://educationdata.urban.org/csv/{file_dir}/{file_name}` |
| `GET /api/v1/api-variables` | Every variable in the portal (data types, labels, special values) |
| `GET /api/v1/api-endpoint-varlist` | Variables broken out by endpoint (unique endpoint_id × variable) |

## IPEDS — Directory & Institutional Characteristics

| Endpoint | Years | Description |
|----------|-------|-------------|
| `.../ipeds/directory/{year}/` | 1980, 1984–2023 | Name, address, state, sector, level, Carnegie classification, HBCU/tribal flags, Title IV status. One row per institution. |
| `.../ipeds/institutional-characteristics/{year}/` | 1980, 1984–2023 | Program offerings, award levels, control/affiliation, special learning opportunities, student services, disability services, tuition plans, athletic conferences, veteran services. |

## Query Tips

- **Filter by institution:** append `?unitid=166027` (colleges), `?ncessch=...` (schools), or `?leaid=...` (districts).
- **Filter by FIPS state code:** append `?fips=25` (Massachusetts = 25).
- **Pagination:** responses include `count`, `next`, and `previous`. Follow `next` URLs to page through large result sets.
- **Multiple filters:** combine with `&`, e.g. `?unitid=166027&sex=99&race=99`.
- **Coded values:** many fields (sex, race, level_of_study, etc.) use integer codes. Consult the `/api-variables` metadata endpoint or the [documentation](https://educationdata.urban.org/documentation/) for value labels.
